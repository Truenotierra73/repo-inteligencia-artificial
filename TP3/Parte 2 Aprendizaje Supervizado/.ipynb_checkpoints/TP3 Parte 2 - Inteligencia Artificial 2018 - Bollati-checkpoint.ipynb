```python
# LIBRERÍAS
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from scipy import stats
import numpy as np

# carga del datasets
data = load_breast_cancer()

# para resolver este ejericio se tomarán todo los features
X = data.data
y = data.target

# Dividimos el conjunto en train, test. El 70% para entrenamiento y el 30% para test.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.3)
# Subdivisión del conjunto anterior. Tomé el 70% anterior para subdividirlo en 55% para entrenamiento y 15% para test.
# Cabe destacar que se realiza la subdivisión sobre la misma variable. Por lo tanto, se "pisan" los datos.
X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train, random_state=10, test_size=0.15)

# A continuación se calcula la media aritmética de cada uno de los predictores.
# Como parámetro se pasa el subconjunto de entrenamiento para la normalización de los datos. Axis = 0 significa
# que recorrerá la matriz por sus columnas.
media_de_features = np.mean(X_train, axis=0)
# A continuación se calcula la desviación estándar de los datos del subconjunto de entrenamiento
# para cada uno de los predictores.
desviacion_de_features = np.std(X_train, axis=0)

# Declaración de la variable que representará la normalización de los datos del subconjunto de entrenamiento.
# np.zeros(np.shape(X_train)) se hace para que se declare con la misma dimensión del subconjunto X_train.
X_train_norm = np.zeros(np.shape(X_train))

# ACLARACIÓN: este comentario es solo para informar que  el código de abajo es solo para realizar la normalización
# de forma manual.
# ----INICIO----
# for i in range(X_train.shape[1]):
#     for j in range(X_train.shape[0]):
#         X_train_norm[j, i] = ((X_train[j, i] - media_de_features[i]) / desviacion_de_features[i])
# ----FIN----

# Cálculo automático de la normalización de X_train
X_train_norm = stats.zscore(X_train, axis=0)

# Resultado
print("Normalización de X_train")
print("--------------------------------")
print(X_train_norm)

# Declaración de la variable que representará la normalización de los datos del subconjunto de validación.
X_validation_norm = np.zeros(np.shape(X_validation))

# Forma manual de realizar la normalización de los datos. Normalización de X_validation.
for i in range(X_validation.shape[1]): # Se inicia recorriento la matriz a través de las columnas.
    for j in range(X_validation.shape[0]): # Se recorre la matriz por la columna i, en la fila j.
        # Cálculo de la normalización
        X_validation_norm[j, i] = ((X_validation[j, i] - media_de_features[i]) / desviacion_de_features[i])

# Resultado
print("\n")
print("Normalización de X_validation")
print("-----------------------------")
print(X_validation_norm)

# Declaración de la variable que representará la normalización de los datos del subconjunto de test.
X_test_norm = np.zeros(np.shape(X_test))

# Normalización manual de X_test
for i in range(X_test.shape[1]):
    for j in range(X_test.shape[0]):
        X_test_norm[j, i] = ((X_test[j, i] - media_de_features[i]) / desviacion_de_features[i])

# Resultado
print("\n")
print("Normalización de X_test")
print("-----------------------")
print(X_test_norm)
```

```python
# LIBRERÍAS
import numpy as np
import matplotlib.pyplot as plt
import scikitplot as skplt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn import neighbors
from sklearn.metrics import accuracy_score
from scipy import stats
from sklearn import metrics

# carga del datasets
data = load_breast_cancer()

# para resolver este ejericio se tomarán todo los features
X = data.data
y = data.target

# VARIABLES
lista_kNN = [] # lista para guardar permanentemente el valor de kNN
lista_tasa_de_aciertos = [] # lista para guardar cada una de las tasas de aciertos en la búsqueda de optimización de k
tasa_de_aciertos = 0 # varieble utilizada para guardar la mejor tasta de aciertos
k = 1 # variable utilizada para optimizar la distancia k de kNN haciendo un grid-search
kNN = 0 # variable utilizada para guardar el mejor kNN

# Dividimos el conjunto en train, test. El 70% para entrenamiento y el 30% para test.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.3)
# Subdivisión del conjunto anterior. Tomé el 70% anterior para subdividirlo en 55% para entrenamiento y 15% para test.
# Cabe destacar que se realiza la subdivisión sobre la misma variable. Por lo tanto, se "pisan" los datos.
X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train, random_state=10, test_size=0.15)

# A continuación se calcula la media aritmética de cada uno de los predictores.
# Como parámetro se pasa el subconjunto de entrenamiento para la normalización de los datos. Axis = 0 significa
# que recorrerá la matriz por sus columnas.
media_de_features = np.mean(X_train, axis=0)
# A continuación se calcula la desviación estándar de los datos del subconjunto de entrenamiento
# para cada uno de los predictores.
desviacion_de_features = np.std(X_train, axis=0)

# Declaración de la variable que representará la normalización de los datos del subconjunto de entrenamiento.
# np.zeros(np.shape(X_train)) se hace para que se declare con la misma dimensión del subconjunto X_train.
X_train_norm = np.zeros(np.shape(X_train))
# Cálculo automático de la normalización de X_train
X_train_norm = stats.zscore(X_train, axis=0)

# Declaración de la variable que representará la normalización de los datos del subconjunto de validación.
X_validation_norm = np.zeros(np.shape(X_validation))
# Forma manual de realizar la normalización de los datos. Normalización de X_validation.
for i in range(X_validation.shape[1]): # Se inicia recorriento la matriz a través de las columnas.
    for j in range(X_validation.shape[0]): # Se recorre la matriz por la columna i, en la fila j.
        # Cálculo de la normalización
        X_validation_norm[j, i] = ((X_validation[j, i] - media_de_features[i]) / desviacion_de_features[i])

# Declaración de la variable que representará la normalización de los datos del subconjunto de test.
X_test_norm = np.zeros(np.shape(X_test))
# Normalización manual de X_test
for i in range(X_test.shape[1]):
    for j in range(X_test.shape[0]):
        X_test_norm[j, i] = ((X_test[j, i] - media_de_features[i]) / desviacion_de_features[i])

# IMPLEMENTACIÓN DEL GRID-SEARCH PARA LA OPTIMIZACIÓN DE k
for k in range(1, 21):
    # Instanciamos el clasificador
    clf_sk = neighbors.KNeighborsClassifier(n_neighbors=k)

    # Entreno el clasificador con los datos normalizados
    clf_sk.fit(X_train_norm, y_train)
    # Cálculo de y_hat con el X_validation_norm
    y_hat = clf_sk.predict(X_validation_norm)

    # Cálculo de la tasa de aciertos
    tasa_de_aciertos = accuracy_score(y_true=y_validation, y_pred=y_hat)

    lista_kNN.append(k) # guardo el k en una lista
    lista_tasa_de_aciertos.append(tasa_de_aciertos) # agregamos la tasa de aciertos a una lista

    # reinicio las variables
    mejor_tasa_de_aciertos = 0

# Obtengo el valor máximo de la la lista de la tasa de aciertos
tasa_de_aciertos = max(lista_tasa_de_aciertos)
# Obtengo la posición (índice) en el cuál se encuentra la tasa de aciertos en la lista para encontrar el kNN correspondiente
kNN_posicion = lista_tasa_de_aciertos.index(tasa_de_aciertos)
# Obtengo el kNN de la lista pasando como parámetro la posición
kNN = lista_kNN[kNN_posicion]

# Vuelvo a instanciar el clasificador, pero con el mejor kNN que maximiza la tasa de aciertos
clf_sk = neighbors.KNeighborsClassifier(n_neighbors=kNN)
# Entreno el clasificador con los datos normalizados
clf_sk.fit(X_train_norm, y_train)
# Cálculo del y_hat, pero con el conjunto X_test_norm
y_hat = clf_sk.predict(X_test_norm)
# Cálculo de la matriz de probabilidades
y_hat_proba = clf_sk.predict_proba(X_test_norm)

# Resultados
print("La tasa de aciertos antes del entrenamiento:", tasa_de_aciertos)
print("La mejor kNN es:", kNN, "y su tasa de aciertos después de entrenarlo es:", accuracy_score(y_true=y_test, y_pred=y_hat))
print("El área bajo la curva es:", metrics.roc_auc_score(y_true=y_test, y_score=y_hat))
print("\n")
# Reporte de clasificación (precision, recall, f1-score y support)
print(metrics.classification_report(y_test, y_hat, target_names=data.target_names))
skplt.metrics.plot_roc_curve(y_test, y_hat_proba) # gráfico
plt.show()
```

```python
# LIBRERÍAS
import numpy as np
import matplotlib.pyplot as plt
import scikitplot as skplt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn import neighbors
from scipy import stats
from sklearn import metrics

# carga del datasets
data = load_breast_cancer()

# para resolver este ejericio se tomarán todo los features
X = data.data
y = data.target

# VARIABLES
costo_falso_positivo = 7 # LFP, representa el costo de un error falso positivo
costo_falso_negativo = 18 # LFN, representa el costo de un error falso negativo

# Dividimos el conjunto en train, test. El 70% para entrenamiento y el 30% para test.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.3)
# Subdivisión del conjunto anterior. Tomé el 70% anterior para subdividirlo en 55% para entrenamiento y 15% para test.
# Cabe destacar que se realiza la subdivisión sobre la misma variable. Por lo tanto, se "pisan" los datos.
X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train, random_state=10, test_size=0.15)

# A continuación se calcula la media aritmética de cada uno de los predictores.
# Como parámetro se pasa el subconjunto de entrenamiento para la normalización de los datos. Axis = 0 significa
# que recorrerá la matriz por sus columnas.
media_de_features = np.mean(X_train, axis=0)
# A continuación se calcula la desviación estándar de los datos del subconjunto de entrenamiento
# para cada uno de los predictores.
desviacion_de_features = np.std(X_train, axis=0)

# Declaración de la variable que representará la normalización de los datos del subconjunto de entrenamiento.
# np.zeros(np.shape(X_train)) se hace para que se declare con la misma dimensión del subconjunto X_train.
X_train_norm = np.zeros(np.shape(X_train))
# Cálculo automático de la normalización de X_train
X_train_norm = stats.zscore(X_train, axis=0)

# Declaración de la variable que representará la normalización de los datos del subconjunto de validación.
X_validation_norm = np.zeros(np.shape(X_validation))
# Forma manual de realizar la normalización de los datos. Normalización de X_validation.
i = 0
for i in range(X_validation.shape[1]): # Se inicia recorriento la matriz a través de las columnas.
    for j in range(X_validation.shape[0]): # Se recorre la matriz por la columna i, en la fila j.
        # Cálculo de la normalización
        X_validation_norm[j, i] = ((X_validation[j, i] - media_de_features[i]) / desviacion_de_features[i])

# Declaración de la variable que representará la normalización de los datos del subconjunto de test.
X_test_norm = np.zeros(np.shape(X_test))
# Normalización manual de X_test
i = 0
for i in range(X_test.shape[1]):
    for j in range(X_test.shape[0]):
        X_test_norm[j, i] = ((X_test[j, i] - media_de_features[i]) / desviacion_de_features[i])

# Instancio el clasificador con un kNN = 3, tomando el mejor k optimizado del ejercicio básico 2.
clf_sk = neighbors.KNeighborsClassifier(n_neighbors=3)
# Entreno el clasificador
clf_sk.fit(X_train_norm, y_train)
# Cálculo de y_hat
y_hat = clf_sk.predict(X_test_norm)
# Cálculo de la matriz de probabilidades
y_hat_proba = clf_sk.predict_proba(X_test_norm)

# Cálculo de la y_hat teniendo en cuenta los costos de error (LFN y LFP)
y_hat_alterada = [] # vector que almacenará los valores de la nueva prediccón considerando los costos
i = 0
# Para realizar el cálculo de y_hat_alterada, se implementó un ciclo for que recorra todas las filas
# de la matriz de probabilidades en la columna 0.
# En la columna 0 se encuentran las probabilidades de las células malignas, mientras que en la columna 1
# se encuentran las probabilidades de las células benignas.
for i in range(y_hat_proba.shape[0]):
    # Cálculo de la probabilidad de elegir maligno (y==0)
    res = ((1 / (costo_falso_negativo / costo_falso_positivo)) * y_hat_proba[i,0]) # "res" significa resultado
    # Condición necesaria para determinar que:
    # Elijo benigno (y==1) si la probabilidad de que lo sea es mayor que la probabilidad que sea maligno (y==0)
    # en caso contrario, elijo maligno
    if y_hat_proba[i,1] > res:
        y_hat_alterada.append(1)
    else:
        y_hat_alterada.append(0)
    # reinicio la variable "res"
    res = 0
# Convierto el vector en un vector de numpy
y_hat_alterada = np.array(y_hat_alterada)

# Resultados
print("y_hat_alterada")
print("--------------")
print(y_hat_alterada)

print("\n")
print("Confusion Matrix y_hat_alterada")
print("-------------------------------")
print(metrics.confusion_matrix(y_true=y_test, y_pred=y_hat_alterada, labels=[0,1]))
skplt.metrics.plot_confusion_matrix(y_true=y_test, y_pred=y_hat_alterada, labels=[0,1])
plt.show()

print("y_hat")
print("-----")
print(y_hat)

print("\n")
print("Confusion Matrix y_hat")
print("----------------------")
print(metrics.confusion_matrix(y_true=y_test, y_pred=y_hat, labels=[0,1]))
skplt.metrics.plot_confusion_matrix(y_true=y_test, y_pred=y_hat, labels=[0,1])
plt.show()
```

```python
# LIBRERÍAS
import numpy as np
import matplotlib.pyplot as plt
import scikitplot as skplt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn import neighbors
from scipy import stats
from sklearn import metrics

# carga del datasets
data = load_breast_cancer()

# para resolver este ejericio se tomarán todo los features
X = data.data
y = data.target

# VARIABLES
k = 1 # variable utilizada para optimizar la distancia k de kNN haciendo un grid-search
kNN = 0 # variable utilizada para guardar el mejor kNN a partir de la lista_mejor_kNN
lista_mejor_kNN = [] # lista utilizada para guardar cada uno de los kNN en el grid-search
lista_menor_costo_total_de_error = [] # lista utilizada para guardar cada uno de los costos calculados en el grid-search
costo_total_de_error = 0 # variable utilizada para derminar el costo total de error y guardarlo en la lista.
costo_falso_positivo = 7 # LFP, representa el costo de un error falso positivo
costo_falso_negativo = 18 # LFN, representa el costo de un error falso negativo
cant_FP = 0 # cantidad de falsos postivos
cant_FN = 0 # cantidad de falsos negativos

# Dividimos el conjunto en train, test. El 70% para entrenamiento y el 30% para test.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.3)
# Subdivisión del conjunto anterior. Tomé el 70% anterior para subdividirlo en 55% para entrenamiento y 15% para test.
# Cabe destacar que se realiza la subdivisión sobre la misma variable. Por lo tanto, se "pisan" los datos.
X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train, random_state=10, test_size=0.15)

# A continuación se calcula la media aritmética de cada uno de los predictores.
# Como parámetro se pasa el subconjunto de entrenamiento para la normalización de los datos. Axis = 0 significa
# que recorrerá la matriz por sus columnas.
media_de_features = np.mean(X_train, axis=0)
# A continuación se calcula la desviación estándar de los datos del subconjunto de entrenamiento
# para cada uno de los predictores.
desviacion_de_features = np.std(X_train, axis=0)

# Declaración de la variable que representará la normalización de los datos del subconjunto de entrenamiento.
# np.zeros(np.shape(X_train)) se hace para que se declare con la misma dimensión del subconjunto X_train.
X_train_norm = np.zeros(np.shape(X_train))
# Cálculo automático de la normalización de X_train
X_train_norm = stats.zscore(X_train, axis=0)

# Declaración de la variable que representará la normalización de los datos del subconjunto de validación.
X_validation_norm = np.zeros(np.shape(X_validation))
# Forma manual de realizar la normalización de los datos. Normalización de X_validation.
i = 0
for i in range(X_validation.shape[1]): # Se inicia recorriento la matriz a través de las columnas.
    for j in range(X_validation.shape[0]): # Se recorre la matriz por la columna i, en la fila j.
        # Cálculo de la normalización
        X_validation_norm[j, i] = ((X_validation[j, i] - media_de_features[i]) / desviacion_de_features[i])

# Declaración de la variable que representará la normalización de los datos del subconjunto de test.
X_test_norm = np.zeros(np.shape(X_test))
# Normalización manual de X_test
i = 0
for i in range(X_test.shape[1]):
    for j in range(X_test.shape[0]):
        X_test_norm[j, i] = ((X_test[j, i] - media_de_features[i]) / desviacion_de_features[i])

# IMPLEMENTACIÓN DEL GRID-SEARCH PARA LA OPTIMIZACIÓN DE k
for k in range(1, 21):
    # Instanciamos el clasificador
    clf_sk = neighbors.KNeighborsClassifier(n_neighbors=k)
    # Entreno el clasificador con los datos normalizados
    clf_sk.fit(X_train_norm, y_train)
    # Cálculo de y_hat_proba con el X_validation_norm
    y_hat_proba = clf_sk.predict_proba(X_validation_norm)

    # Cálculo de la y_hat teniendo en cuenta los costos de error (LFN y LFP)
    y_hat_alterada = []
    i = 0
    # Para realizar el cálculo de y_hat_alterada, se implementó un ciclo for que recorra todas las filas
    # de la matriz de probabilidades en la columna 0.
    # En la columna 0 se encuentran las probabilidades de las células malignas, mientras que en la columna 1
    # se encuentran las probabilidades de las células benignas.
    for i in range(y_hat_proba.shape[0]):
        # Cálculo de la probabilidad de elegir maligno (y==0)
        res = ((1 / (costo_falso_negativo / costo_falso_positivo)) * y_hat_proba[i,0]) # "res" significa resultado
        # Condición necesaria para determinar que:
        # Elijo benigno (y==1) si la probabilidad de que lo sea es mayor que la probabilidad que sea maligno (y==0)
        # en caso contrario, elijo maligno
        if y_hat_proba[i,1] > res:
            y_hat_alterada.append(1)
        else:
            y_hat_alterada.append(0)
        # reinicio la variable "res"
        res = 0
    # Convierto el vector en un vector de numpy
    y_hat_alterada = np.array(y_hat_alterada)

    # Realizo la comparación entre el y_hat_alterada y el y_validation para determinar la cantidad de falsos positivos
    # y la cantidad de falsos negativos
    i = 0
    for i in range(0, len(y_validation)):
        # Condición necesaria para determinar que:
        # Si el valor del y_hat_alterada es maligno (y==0) y en realidad es benigno (y==1), entonces es un falso positivo.
        # Mientras que, si el valor del y_hat_alterada es benigno y en realidad es maligno, entonces es un falso negativo.
        if y_hat_alterada[i] == 0 and y_validation[i] == 1:
            cant_FP += 1
        elif y_hat_alterada[i] == 1 and y_validation[i] == 0:
            cant_FN += 1

    costo_total_de_error = (costo_falso_positivo * cant_FP) + (costo_falso_negativo * cant_FN) # calculo el costo total de error
    lista_menor_costo_total_de_error.append(costo_total_de_error) # guardo el costo total de error en una lista
    lista_mejor_kNN.append(k)  # guardo el k en una lista

    # reinicio las variables
    costo_total_de_error = 0
    y_hat_alterada = []
    cant_FN = 0
    cant_FP = 0

# Cálculo del menor costo total de error junto con el kNN correspondiente.
menor_costo_total_de_error = min(lista_menor_costo_total_de_error) # obtengo el mínimo costo total de error
kNN_posicion = lista_menor_costo_total_de_error.index(menor_costo_total_de_error) # obtengo la posición del mínimo costo total de error
kNN = lista_mejor_kNN[kNN_posicion] # obtengo el valor kNN de la posición del valor mínimo

# resultados
print("El menor costo total de error es:" , menor_costo_total_de_error)
print("El mejor kNN es:" , kNN)
print("\n")

# Instancio el calsificador con el mejor kNN encontrado
clf_sk = neighbors.KNeighborsClassifier(n_neighbors=kNN)
# Entreno el clasificador
clf_sk.fit(X_train_norm, y_train)
# Cálculo de la matriz de probabilidades
y_hat_proba = clf_sk.predict_proba(X_validation_norm)

y_hat_alterada = []
i = 0
for i in range(y_hat_proba.shape[0]):
    res = ((1 / (costo_falso_negativo / costo_falso_positivo)) * y_hat_proba[i,0])
    if y_hat_proba[i,1] > res:
        y_hat_alterada.append(1)
    else:
        y_hat_alterada.append(0)
    res = 0
y_hat_alterada = np.array(y_hat_alterada)

# Confusion matrix y un gráfico a modo de visualizar lo realizado.
# Es una forma sencilla de evaluar el rendimiento de un algoritmo es por medio de una confusion matrix o error matrix, 
# la cual contrasta las predicciones con los valores verdaderos, mostrando los errores TI (Tipo 1) y TII (Tipo 2).
print(metrics.confusion_matrix(y_true=y_validation, y_pred=y_hat_alterada, labels=[0,1]))
skplt.metrics.plot_confusion_matrix(y_true=y_validation, y_pred=y_hat_alterada, labels=[0,1])
plt.show()
```
