## Ejercicio 1.1

```python
# LIBRERÍAS
import numpy as np
import matplotlib.pyplot as plt
from sklearn import neighbors
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# CARGO EL DATASET
iris = load_iris()

# tomamos los primeros dos features para mostrarlos mejor gráficamente (el largo y el ancho del sépalo)
# (por eso el iris.data[:, :2])
X = iris.data[:, :2]
y = iris.target


class ClasificadorNN(object):  # <object> en [...]_NN(object) indica que hereda de la clase object

    # <self> indica referencia a un objeto concreto de la clase
    def __init__(self):
        # inicializamos ambos atributos X e y arbitrariamente
        self.X = []
        self.y = []

    def fit(self, X_train, y_train):
        # para el caso de KNN, nuestro entrenamiento sólo consiste en recordar los datos
        # (X, y) de entrenamiento
        self.X = X_train
        self.y = y_train

    def predict(self, X_test):
        # inicializamos nuestra variable de salida, y_hat, con un tamaño igual a la
        # cantidad de filas
        y_hat = np.zeros(X_test.shape[0])

        for i in range(X_test.shape[0]):  # vamos a iterar sobre todas las filas de datos

            # ELEVAMOS AL CUADRADO LA MATRIZ DISTANCIA PARA IMPLEMENTAR LA DISTANCIA EUCLIDEANA
            matriz_distancia = np.power(self.X - X_test[i, :], 2)
            # LUEGO, APLICO RAIZ CUADRADA A LA SUMA DE LA MATRIZ DISTANCIA PARA TERMINAR DE APLICAR LA DISTANCIA EUCLIDEANA
            suma_distancias = np.sqrt(np.sum(matriz_distancia, axis=1))

            indice_punto_mas_cercano = np.argmin(suma_distancias)
            y_hat[i] = self.y[indice_punto_mas_cercano]

        return y_hat


# Dividimos el conjunto en train, test. El 70% para entrenamiento y el 30% para test.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.3)

# Instanciamos nuestro clasificador
clf_sk = neighbors.KNeighborsClassifier(n_neighbors=1)

# entrenamos el modelo
clf_sk.fit(X_train,y_train)
# predecimos la salida
y_hat = clf_sk.predict(X_test)

plt.scatter(X_test[y_test==0, 0], X_test[y_test==0, 1], label=iris.target_names[0], color='orange')
plt.scatter(X_test[y_test==1, 0], X_test[y_test==1, 1], label=iris.target_names[1], color='blue')
plt.scatter(X_test[y_test==2, 0], X_test[y_test==2, 1], label=iris.target_names[2], color='green')
plt.scatter(X_test[y_test!=y_hat, 0], X_test[y_test!=y_hat, 1], label='clasif. incorrecta', c='black') # Agregamos las malas predicciones

plt.legend(loc='best', numpoints=1)
plt.xlabel('Largo del sépalo')
plt.ylabel('Ancho del sépalo')

# TASA DE ACIERTOS
print('Tasa de aciertos: ', accuracy_score(y_true=y_test, y_pred=y_hat))

# MUESTRO EL GRÁFICO
plt.show()
```

## Ejercicio 1.3

```python
# LIBRERÍAS
import numpy as np
import matplotlib.pyplot as plt
from sklearn import neighbors
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# CARGO EL DATASET
iris = load_iris()

# MODIFICACIÓN EN LOS PREDICTORES QUE TOMO DEL DATASET
# tomamos los últimos dos features del dataset para mostrarlos mejor gráficamente (el largo y el ancho del pétalo)
# (por eso el iris.data[:, 2:5]) ó iris.data[:, 2:} en este caso porque tenemos 4 predictores, sino tenemos que especificar desde-hasta en las columnas que tomamos.
X = iris.data[:, 2:4]
y = iris.target

class ClasificadorNN(object):  # <object> en [...]_NN(object) indica que hereda de la clase object

    # <self> indica referencia a un objeto concreto de la clase
    def __init__(self):
        # inicializamos ambos atributos X e y arbitrariamente
        self.X = []
        self.y = []

    def fit(self, X_train, y_train):
        # para el caso de KNN, nuestro entrenamiento sólo consiste en recordar los datos
        # (X, y) de entrenamiento
        self.X = X_train
        self.y = y_train

    def predict(self, X_test):
        # inicializamos nuestra variable de salida, y_hat, con un tamaño igual a la
        # cantidad de filas
        y_hat = np.zeros(X_test.shape[0])

        for i in range(X_test.shape[0]):  # vamos a iterar sobre todas las filas de datos
            vector_distancia = np.abs(self.X - X_test[i, :])

            # axis=1 significa que la suma se realiza entre las columnas (axis=0
            # haría que la suma se realiza entre las filas). axis=None haría que
            # la suma se realize entre todos los elementos, devolviendo un escalar
            # Para más info ver sección Examples en
            # https://docs.scipy.org/doc/numpy/reference/generated/numpy.sum.html
            suma_distancias = np.sum(vector_distancia, axis=1)
            indice_punto_mas_cercano = np.argmin(suma_distancias)
            y_hat[i] = self.y[indice_punto_mas_cercano]

        return y_hat

# Dividimos el conjunto en train, test. El 70% para entrenamiento y el 30% para test.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.3)

# Instanciamos nuestro clasificador
clf_sk = neighbors.KNeighborsClassifier(n_neighbors=1)

# entrenamos el modelo
clf_sk.fit(X_train,y_train)
# predecimos la salida
y_hat = clf_sk.predict(X_test)

plt.scatter(X_test[y_test==y_hat, 0], X_test[y_test==y_hat, 1], label='clasif. correcta', color='green') # CLASIFICACIÓN CORRECTA SIN IMPORTAR LAS CLASES
plt.scatter(X_test[y_test!=y_hat, 0], X_test[y_test!=y_hat, 1], label='clasif. incorrecta', c='red') # CLASIFICACIÓN INCORRECTA (MALAS PREDICCIONES)

plt.legend(loc='best', numpoints=1)
plt.xlabel('Largo del pétalo')
plt.ylabel('Ancho del pétalo')

# TASA DE ACIERTOS
print('Tasa de aciertos: ', accuracy_score(y_true=y_test, y_pred=y_hat))

# MUESTRO EL GRÁFICO
plt.show()
```

## Ejercicio 2.1

```python
# LIBRERÍAS
import numpy as np
import matplotlib.pyplot as plt
from sklearn import neighbors
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold

# CARGO EL DATASET
iris = load_iris()


X = iris.data[:, :2] # utilicé todas las observaciones, pero solo los dos primeros predictores (Largo y ancho del sépalo)
y = iris.target

# VARIABLES
suma_aciertos = 0 # es la variable utilizada como acumulador de todas las tasas de aciertos
i = 0 # variable utilizada para solo imprimir las tasas de aciertos para un deterinado kf

# Dividimos el conjunto en train, test. El 70% para entrenamiento y el 30% para test.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.3)

# n_splits es el número de divisiones, las cuales
kf = KFold(n_splits=5, shuffle=True, random_state=40)

# El siguiente for nos abstrae de tener que manejar explícitamente los subíndices de los folds,
# al dividir el conjunto de entrenamiento iterativamente. En cada una de sus iteraciones
# obtenemos los subíndices de todos los folds de entrenamiento y del fold de validación de una
# división en particular, de modo tal que todos los k folds sean al menos una vez usados para
# entrenar el modelo y al menos una vez usados para validar los híper-parámetros del modelo
for train_index, validation_index in kf.split(X_train):
    X_train_kf, X_validation_kf = X_train[train_index], X_train[validation_index]
    y_train_kf, y_validation_kf = y_train[train_index], y_train[validation_index]

    # desde aquí estamos trabajando con una división train-validation en particular
    # es decir, por ejemplo, que train_index nos da todos los índices que componen a
    # los folds que se usarán para entrenar el modelo en esta división. Usamos estos
    # índices para obtener los X_train_kf y los y_train_kf. Lo mismo sucede con los
    # índices usados para el fold de validación en esta división. Si se desea ver
    # cómo cambian los índices en cada iteración, hacer un print(validation_index)

    i += 1

    # Instanciamos nuestro clasificador
    clf_sk = neighbors.KNeighborsClassifier(n_neighbors=1)

    # entrenamos el modelo
    clf_sk.fit(X_train_kf,y_train_kf)
    # predecimos la salida
    y_hat = clf_sk.predict(X_validation_kf)

    # TASA DE ACIERTOS DE CADA K FOLD
    print('Tasa de aciertos para K Fold N°', i, ':', accuracy_score(y_true=y_validation_kf, y_pred=y_hat))
    suma_aciertos = suma_aciertos + accuracy_score(y_true=y_validation_kf, y_pred=y_hat) # acumulador de la tasa de aciertos para cada k fold

# CÁLCULO DEL PROMEDIO DE LA TASA DE ACIERTOS
promedio_tasa_aciertos = suma_aciertos/kf.n_splits
# MUESTRO EL RESULTADO POR PANTALLA
print('Promedio de Tasa de Aciertos:', promedio_tasa_aciertos)
```

Utilizando el Largo y Ancho del Sépalo podemos observar que las **tasas de aciertos** para cada K Fold tienen diferencias considerables comparándolas entre cada una de ellas. La tasa de aciertos más alta se puede ver en el K Fold N° 3. Al dividir el dataset primero en 70% para entrenamiento, mezclar los datos y luego realizar 5 subdivisiones para realizar una validación cruzada, puede ser posible que la diferencia de resultados dependa de la mezcla de los datos, la cantidad de datos que se utilice, la cantidad de predictores utilizada o la semilla de utilizada de aleatorización para la mezcla de los datos.

```python
# LIBRERÍAS
import numpy as np
import matplotlib.pyplot as plt
from sklearn import neighbors
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold

# CARGO EL DATASET
iris = load_iris()

X = iris.data # para la realización de este ejercicio se tomó todo el dataset
y = iris.target

# VARIABLES
n_kf = 0 # variable utilizada para solo imprimir las tasas de aciertos para un deterinado k Fold
k = 1 # variable utilizada para optimizar la distancia k de kNN haciendo un grid-search
kNN = 0 # variabla utilizada para guardar parcialmente el mejor kNN para un determinado K Fold
lista_mejor_kNN = [] # lista para guardar permanentemente el mejor valor de kNN
lista_mejor_tasa_de_aciertos = [] # lisata para guardar permanentemente la mejor tasa de aciertos de un determinado K Fold y para un k en particular
tasa_de_aciertos = 0 # variable utilizada para guardar parcialmente la mejor tasa de aciertos para un determinado k
mejor_tasa_de_aciertos = 0 # variable utilizada para guardar la mejor tasa de aciertos permanentemente

# Dividimos el conjunto en train, test. El 70% para entrenamiento y el 30% para test.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.3)

# n_splits es el número de divisiones, las cuales
kf = KFold(n_splits=5, shuffle=True, random_state=40)

# El siguiente for nos abstrae de tener que manejar explícitamente los subíndices de los folds,
# al dividir el conjunto de entrenamiento iterativamente. En cada una de sus iteraciones
# obtenemos los subíndices de todos los folds de entrenamiento y del fold de validación de una
# división en particular, de modo tal que todos los k folds sean al menos una vez usados para
# entrenar el modelo y al menos una vez usados para validar los híper-parámetros del modelo
for train_index, validation_index in kf.split(X_train):
    X_train_kf, X_validation_kf = X_train[train_index], X_train[validation_index]
    y_train_kf, y_validation_kf = y_train[train_index], y_train[validation_index]

    # desde aquí estamos trabajando con una división train-validation en particular
    # es decir, por ejemplo, que train_index nos da todos los índices que componen a
    # los folds que se usarán para entrenar el modelo en esta división. Usamos estos
    # índices para obtener los X_train_kf y los y_train_kf. Lo mismo sucede con los
    # índices usados para el fold de validación en esta división. Si se desea ver
    # cómo cambian los índices en cada iteración, hacer un print(validation_index)

    # IMPLEMENTACIÓN DEL GRID-SEARCH PARA LA OPTIMIZACIÓN DE k
    for k in range(1,21):
        # Instanciamos nuestro clasificador
        clf_sk = neighbors.KNeighborsClassifier(n_neighbors=k)

        # entrenamos el modelo
        clf_sk.fit(X_train_kf,y_train_kf)
        # predecimos la salida
        y_hat = clf_sk.predict(X_validation_kf)

        # CÁLCULO DE TASA DE ACIERTOS
        tasa_de_aciertos = accuracy_score(y_true=y_validation_kf, y_pred=y_hat)
        if tasa_de_aciertos > mejor_tasa_de_aciertos:
            mejor_tasa_de_aciertos = tasa_de_aciertos
            kNN = k

    lista_mejor_kNN.append(kNN) # agregamos el mejor k a la lista para un determinado K Fold
    lista_mejor_tasa_de_aciertos.append(mejor_tasa_de_aciertos) # agregamos la mejor tasa de aciertos para un determinado K Fold
    
    # reinicio las variables
    mejor_tasa_de_aciertos = 0
    kNN = 0

# LOOP IMPLEMENTADO PARA IMPRIMIR LAS MEJORES TASAS DE ACIERTOS PARA CADA UNO DE LOS K FOLDS Y PARA UN DETERMINADO k OPTIMIZADO
for n_kf in range(n_kf, lista_mejor_tasa_de_aciertos.__len__()):
    print('La mejor tasa de aciertos para K Fold N°', n_kf + 1, 'es', lista_mejor_tasa_de_aciertos[n_kf], 'para kNN =', lista_mejor_kNN[n_kf])
```

## Ejercicio 2.3

```python
# LIBRERÍAS
import numpy as np
import matplotlib.pyplot as plt
from sklearn import neighbors
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold

# CARGO EL DATASET
iris = load_iris()

X = iris.data # para la realización de este ejercicio se tomó todo el dataset
y = iris.target

# VARIABLES
n_kf = 0 # variable utilizada para solo imprimir las tasas de aciertos para un deterinado k Fold
k = 1 # variable utilizada para optimizar la distancia k de kNN haciendo un grid-search
kNN = 0 # variabla utilizada para guardar parcialmente el mejor kNN para un determinado K Fold
lista_mejor_kNN = [] # lista para guardar permanentemente el mejor valor de kNN
tasa_de_aciertos = 0 # variable utilizada para guardar parcialmente la mejor tasa de aciertos para un determinado k
mejor_tasa_de_aciertos = 0 # variable utilizada para guardar la mejor tasa de aciertos permanentemente

# Dividimos el conjunto en train, test. El 70% para entrenamiento y el 30% para test.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.3)

# n_splits es el número de divisiones, las cuales
kf = KFold(n_splits=5, shuffle=True, random_state=40)

# El siguiente for nos abstrae de tener que manejar explícitamente los subíndices de los folds,
# al dividir el conjunto de entrenamiento iterativamente. En cada una de sus iteraciones
# obtenemos los subíndices de todos los folds de entrenamiento y del fold de validación de una
# división en particular, de modo tal que todos los k folds sean al menos una vez usados para
# entrenar el modelo y al menos una vez usados para validar los híper-parámetros del modelo
for train_index, validation_index in kf.split(X_train):
    X_train_kf, X_validation_kf = X_train[train_index], X_train[validation_index]
    y_train_kf, y_validation_kf = y_train[train_index], y_train[validation_index]

    # desde aquí estamos trabajando con una división train-validation en particular
    # es decir, por ejemplo, que train_index nos da todos los índices que componen a
    # los folds que se usarán para entrenar el modelo en esta división. Usamos estos
    # índices para obtener los X_train_kf y los y_train_kf. Lo mismo sucede con los
    # índices usados para el fold de validación en esta división. Si se desea ver
    # cómo cambian los índices en cada iteración, hacer un print(validation_index)

    # IMPLEMENTACIÓN DEL GRID-SEARCH PARA LA OPTIMIZACIÓN DE k
    for k in range(1,21):
        # Instanciamos nuestro clasificador
        clf_sk = neighbors.KNeighborsClassifier(n_neighbors=k)

        # entrenamos el modelo
        clf_sk.fit(X_train_kf,y_train_kf)
        # predecimos la salida
        y_hat = clf_sk.predict(X_validation_kf)

        # TASA DE ACIERTOS
        tasa_de_aciertos = accuracy_score(y_true=y_validation_kf, y_pred=y_hat)
        if tasa_de_aciertos > mejor_tasa_de_aciertos:
            mejor_tasa_de_aciertos = tasa_de_aciertos
            kNN = k

    lista_mejor_kNN.append(kNN) # guardo el mejor valor de k en una lista
    
    # reinicio las variables
    mejor_tasa_de_aciertos = 0
    kNN = 0

# UTILIZANDO LOS VALORES OPTIMIZADOS DE k PARA CALCULAR LA TASA DE ACIERTOS AL PREDECIR LOS X DEL CONJUNTO DE TEST
for n_kf in range(n_kf ,5):

    clf_sk = neighbors.KNeighborsClassifier(n_neighbors=lista_mejor_kNN[n_kf])

    clf_sk.fit(X_train, y_train)
    y_hat = clf_sk.predict(X_test)

    # TASA DE ACIERTOS PARA CADA K FOLD
    print('Tasa de aciertos para el K Fold N°', n_kf + 1 , 'es de:', accuracy_score(y_true=y_test, y_pred=y_hat))
```
